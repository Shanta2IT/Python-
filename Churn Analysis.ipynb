```python
missing = pd.DataFrame((pd.read_csv('Customer.csv').isnull().sum())*100/(pd.read_csv('Customer.csv')).shape[0]).reset_index()
plt.figure(figsize=(16,5))
ax = sns.pointplot('index',0,data=missing)
plt.xticks(rotation =90,fontsize =7)
plt.title("Percentage of missing values")
plt.ylabel("PERCENTAGE")
plt.show
```

```python
#To get the pie chart to analyze churn
(pd.read_csv('Customer.csv'))['Churn'].value_counts().plot.pie(explode=[0.05,0.05], autopct='%1.1f%%',  startangle=90,shadow=True, figsize=(8,8))
plt.title('Pie Chart for Churn')
plt.show()
```

```python
#To get the Donut Plot to analyze churn
data = (pd.read_csv('Customer.csv')) ['Churn'].value_counts()
explode = (0, 0.2)
plt.pie(data, explode = explode,autopct='%1.1f%%',shadow=True,radius = 2.0, labels = ['Not churned customer','Churned customer'],colors=['royalblue' ,'lime'])
circle = plt.Circle( (0,0), 1, color='white')
p=plt.gcf()
p.gca().add_artist(circle)
plt.title('Donut Plot for Churn')
plt.show()
```

```python
#let's see churn by using countplot
sns.countplot(x= (pd.read_csv('Customer.csv')).Churn)
```

```python
sns.set(style="darkgrid")
plt.figure(figsize=(16,8))
ax =sns.countplot(x='MonthlyCharges',hue="Churn", data=(pd.read_csv('Customer.csv')))
plt.show()
```

```python
#Analysing by using boxplot diagram between 'MonthlyCharges' and 'churn'
fig = plt.figure(figsize =(10, 8)) 
(pd.read_csv('Customer.csv')).boxplot(column='MonthlyCharges', by='Churn')
fig.suptitle('MonthlyCharges', fontsize=14, fontweight='bold')
plt.show()
```
